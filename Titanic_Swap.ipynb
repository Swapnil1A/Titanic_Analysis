```python
plt.figure(figsize=(5,5))
y=sns.countplot(x='Survived',data=df,hue='Sex')
for bars in y.containers:
    y.bar_label(bars)
plt.title('Survival Distribution by Sex')
plt.show()
```

```python
plt.figure(figsize=(5,5))
s=sns.countplot(x='Pclass',hue='Sex',data=df)
for bars in s.containers:
    s.bar_label(bars)
plt.title('Sex Distribution by Class')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.histplot(df['Age'],bins=30, kde=True)
plt.xlabel('Age')
plt.title('Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.boxplot(x='Pclass',y='Fare',data=df)
plt.title('Fare Distribution by Class')
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.pie(df.Survived.value_counts(),autopct='%.2f%%',labels=['Not Survived(0)','Survived(1)'])
```
